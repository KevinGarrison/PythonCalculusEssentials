```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Wedge
from matplotlib.patches import Polygon
```

### How to Derive and Visualize the Area Formula for a Circle

```python
# Parameters for the circle
center = (0, 0)  # Center of the circle
radius = 3       # Total radius of the circle
dr = 0.5         # Width of each slice
gap = 0.02       # Small gap between slices
slices_to_show = [0.5, 1.5, 2.5]  # Radii of slices to show

# Create a figure and axis
fig, ax = plt.subplots(figsize=(6, 6))

# Set black canvas
fig.patch.set_facecolor('black')
ax.set_facecolor('black')


circle = plt.Circle(center, 3 , color='cyan', alpha=0.7, fill=True)
ax.add_artist(circle) 

# Set limits and aspect
ax.set_xlim(-radius - 1, radius + 1)
ax.set_ylim(-radius - 1, radius + 1)
ax.set_aspect('equal')

# Enable the grid
ax.grid(color='white', linestyle='--', linewidth=0.5)
ax.tick_params(colors='white')  # Make tick labels white for visibility
radius_length = 3
ax.plot([0, radius_length], [0, 0], color='red', linestyle='-', linewidth=2, label='r=3')
# Add title
ax.set_title(f"Circle Area PI*r^2 = {round(np.pi * np.square(radius),2)}", color='white')
plt.legend()
# Show the plot
plt.show()
```

We break it down to a simple example A = 2 * PI * r  for r = 3.

First we slice the Circle into smaller pieces for simplicity with width dr=0.1.

```python
# Parameters for the circles
center = (0, 0)  # Center of the circles
radius_start = 0  # Starting radius
dr = 0.1  # Width of each circle
num_circles = 30  # Number of circles

# Create a figure and axis
fig, ax = plt.subplots(figsize=(6, 6))

# Set black canvas
fig.patch.set_facecolor('black')
ax.set_facecolor('black')

# Loop to draw 5 concentric slices
for i in range(num_circles):
    outer_radius = radius_start + (i + 1) * dr
    inner_radius = radius_start + i * dr
    wedge = Wedge(center, outer_radius, 0, 360, width=dr, color='cyan', alpha=0.7)
    ax.add_patch(wedge)

# Set limits and aspect
ax.set_xlim()
ax.set_ylim(-radius_start - num_circles * dr - 1, radius_start + num_circles * dr + 1)
ax.set_aspect('equal')
radius_length = 3
ax.plot([1.5, 1.6], [0, 0], color='red', linestyle='-', linewidth=2, label='dr=0.1')
# Add title
ax.set_title(f"Concentric Circles with Width {dr}", color='white')
ax.set_xlim(-radius - 1, radius + 1)
ax.set_ylim(-radius - 1, radius + 1)
# Enable the grid
ax.grid(color='white', linestyle='--', linewidth=0.5)
ax.tick_params(colors='white')  # Make tick labels white for visibility
plt.legend()
# Show the plot
plt.show()
```

```python
# Parameters for the circles
dr = 0.1  # Width of each circle slice (will be the width of rectangles)
num_circles = 30  # Number of slices
radius_start = 0  # Starting radius

# Create a figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Set black canvas
fig.patch.set_facecolor('black')
ax.set_facecolor('black')

# Loop to draw rectangles
for i in range(num_circles):
    # Calculate the height of the rectangle (proportional to the circumference of the slice)
    radius = radius_start + (i + 1) * dr
    rect_height = 2 * 3.14159 * radius  # Circumference = 2πr
    rect_x = i * dr  # Position each rectangle horizontally
    rect_y = 0  # Start all rectangles at y=0

    # Draw the rectangle
    rectangle = plt.Rectangle((rect_x, rect_y), dr, rect_height, color='cyan', alpha=0.7)
    ax.add_patch(rectangle)

# Set limits and aspect
ax.set_xlim(0, num_circles * dr)
ax.set_ylim(0, 2 * 3.14159 * (radius_start + num_circles * dr) + 3)
ax.set_aspect(0.1)

# Add lines
height = 2 * 3.14159 * radius
ax.plot([0, radius], [height, height], color='green', linestyle='-', linewidth=2, label='circumference=2πr')
ax.plot([2.9, 3.0], [10, 10], color='red', linestyle='-', linewidth=2, label='dr=0.1')

# Enclosing triangle vertices
triangle_vertices = [
    (0, 0),                   # Bottom-left
    (radius, height),         # Top vertex
    (num_circles * dr, 0),    # Bottom-right
    (0, 0)                    # Close the triangle
]

# Draw the triangle
triangle = Polygon(triangle_vertices, closed=True, edgecolor='yellow', fill=None, linewidth=2, label='Triangle with A = 1/2*b*h with b = 3 and h = 2*PI*r')
ax.add_patch(triangle)

# Add title
ax.set_title(f"Transformed Circle Slices as Rectangles (Width {dr})", color='white')

# Enable grid
ax.grid(color='white', linestyle='--', linewidth=0.5)
ax.tick_params(colors='white')  # Make tick labels white for visibility

# Add legend
plt.legend()

# Show the plot
plt.show()
```

### Derivation of the Area of the circle

The area of a circle can be derived as follows:

$$
\frac{1}{2} \times R \times (2 \pi R) = \pi R^2
$$

```python
# Parameters
radii = np.linspace(0, 3, 30)  # Radii from 0 to 10
circumference = 2 * np.pi * radii  # Circumference formula: 2πr

# Plot
plt.figure(figsize=(8, 5))
plt.plot(radii, circumference, label=r'$C = 2\pi r$', color='cyan', linewidth=2)

# Set black background
plt.gca().set_facecolor('black')  # Set background of the plot area
plt.gcf().set_facecolor('black')  # Set background of the figure

# Grid and axes
plt.grid(True, linestyle='--', linewidth=0.5, color='white')
plt.xlabel("Radius (r)", fontsize=12, color='white')
plt.ylabel("Circumference (C)", fontsize=12, color='white')
plt.title("Graph of Circumference vs Radius", fontsize=14, color='white')
plt.tick_params(colors='white')  # Change tick colors to white

# Legend
plt.legend(fontsize=12, loc='upper left', facecolor='black', edgecolor='white', labelcolor='white')

# Show the plot
plt.show()
```

```python
areas = np.pi * radii**2
# Plot with a black background
plt.figure(figsize=(8, 5))
plt.plot(radii, areas, label=r'$A = \pi r^2$', color='cyan', linewidth=2)

# Set black background
plt.gca().set_facecolor('black')  # Background of the plot area
plt.gcf().set_facecolor('black')  # Background of the entire figure

# Grid and labels
plt.grid(True, linestyle='--', linewidth=0.5, color='white')
plt.xlabel("Radius (r)", fontsize=12, color='white')
plt.ylabel("Area (A)", fontsize=12, color='white')
plt.title("Graph of Area vs Radius", fontsize=14, color='white')
plt.tick_params(colors='white')  # Tick labels in white

# Legend
plt.legend(fontsize=12, loc='upper left', facecolor='black', edgecolor='white', labelcolor='white')

# Show the plot
plt.show()
```

The area under the graph of:

$$
C = 2\pi r
$$

(Circumference vs. Radius) represents the area of a circle because:

1. The circumference:

$$
C = 2\pi r
$$

is the derivative of the area of a circle:

$$
A = \pi r^2
$$

with respect to the radius  r:

$$
\frac{dA}{dr} = 2\pi r
$$

2. Integrating the circumference with respect to \( r \) gives the area of the circle:

$$
A = \int_0^R C(r) \, dr = \int_0^R 2\pi r \, dr
$$

3. Performing the integration:

$$
A = 2\pi \int_0^R r \, dr = 2\pi \left[\frac{r^2}{2}\right]_0^R = \pi R^2
$$
