```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function and its derivative
def s(t):
    return t**3

def ds_dt(t):
    return 3 * t**2

# Parameters for the plot
x = np.linspace(1, 4, 500)  # x values for the function
y = s(x)  # y values for the function
tangent_point = 2  # Point at which to calculate the tangent

# Calculate the tangent line
tangent_slope = ds_dt(tangent_point)
tangent_intercept = s(tangent_point) - tangent_slope * tangent_point
tangent_line = tangent_slope * x + tangent_intercept

# Create the plot
plt.figure(figsize=(8, 6))

# Plot the function
plt.plot(x, y, label=r'$s(t) = t^3$', color='cyan', linewidth=2)

# Plot the tangent line
plt.plot(x, tangent_line, label=r'Tangent at $t=2$', color='green', linewidth=2)

# Set black background
plt.gca().set_facecolor('black')
plt.gcf().set_facecolor('black')

# Grid and labels
plt.grid(True, linestyle='--', linewidth=0.5, color='white')
plt.xlabel("t", fontsize=12, color='white')
plt.ylabel("s(t)", fontsize=12, color='white')
plt.title("Function $s(t) = t^3$ with Tangent at $t=2$", fontsize=14, color='white')
plt.tick_params(colors='white')

# Set limits
plt.xlim(1, 4)
plt.ylim(0, 15)

# Add legend
plt.legend(fontsize=10, loc='upper left', facecolor='black', edgecolor='white', labelcolor='white')

# Show the plot
plt.show()
```

The derivative of s(t) = t^3 at t = 2 can be approximated as:

$$
\frac{ds}{dt}(2) = \lim_{\Delta t \to 0} \frac{(2 + \Delta t)^3 - 2^3}{\Delta t}
$$

Expanding (2 + delta t)^3, we get:

$$
\frac{ds}{dt}(2) = \lim_{\Delta t \to 0} \frac{8 + 12\Delta t + 6\Delta t^2 + \Delta t^3 - 8}{\Delta t}
$$

Simplifying:

$$
\frac{ds}{dt}(2) = \lim_{\Delta t \to 0} \frac{12\Delta t + 6\Delta t^2 + \Delta t^3}{\Delta t}
$$

Factoring delta t:

$$
\frac{ds}{dt}(2) = \lim_{\Delta t \to 0} \left( 12 + 6\Delta t + \Delta t^2 \right)
$$

As delta t -> 0 , the higher-order terms vanish:

$$
\frac{ds}{dt}(2) = 12
$$


That describes an approximation on the rate of change "around this time t"
